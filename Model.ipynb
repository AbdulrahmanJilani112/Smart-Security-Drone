```python
# كشف حالة المنطقة (ازدحام أو طبيعي) باستخدام Logistic Regression
import numpy as np
from sklearn.linear_model import LogisticRegression

# بيانات: [مساحة المكان, عدد الأشخاص] مع التصنيف (0 = طبيعي, 1 = ازدحام)
X = np.array([
    [30, 100],   # طبيعي
    [80, 100],   # ازدحام
    [20, 50],    # طبيعي
    [70, 50],    # ازدحام
])
y = np.array([0, 1, 0, 1])

# تدريب النموذج
clf = LogisticRegression()
clf.fit(X, y)

# اختبار منطقة جديدة
new_area = np.array([[60, 80]])  # 60 شخص في مساحة 80
prediction = clf.predict(new_area)

print("نتيجة النموذج (0 = طبيعي, 1 = ازدحام):", prediction[0])
```

```python
# تصنيف سلوك القيادة إلى آمن أو خطر باستخدام Logistic Regression
import numpy as np
from sklearn.linear_model import LogisticRegression

# بيانات: [الإنحراف, السرعة] مع التصنيف (0 = آمن, 1 = خطر)
X = np.array([
    [50, 2],   # آمن
    [120, 10], # خطر
    [80, 3],   # آمن
    [130, 12], # خطر
])
y = np.array([0, 1, 0, 1])

# تدريب النموذج
clf = LogisticRegression()
clf.fit(X, y)

# اختبار سائق جديد
new_driver = np.array([[100, 8]])
prediction = clf.predict(new_driver)

print("نتيجة النموذج (0 = آمن, 1 = خطر):", prediction[0])
```

```python
# تتبع موقع الهارب باستخدام Linear Regression
import numpy as np
from sklearn.linear_model import LinearRegression

# بيانات: الزمن (ثواني) والموقع X للهارب
time = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
position_x = np.array([2, 4, 6, 8, 10])  # حركة مستقيمة

# تدريب النموذج
model = LinearRegression()
model.fit(time, position_x)

# توقع الموقع عند الزمن 6
predicted_x = model.predict([[6]])
print(" الموقع المتوقع للهارب عند الزمن 6:", predicted_x[0])
```
